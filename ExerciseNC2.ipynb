# **Bài toán nâng cao 2 (NC2)**

```python
# Xác suất đã cho
P_F1 = 0.01
P_F0 = 1 - P_F1

P_L1_given_F1 = 0.9
P_L1_given_F0 = 0.1

P_T1_given_F1 = 0.8
P_T1_given_F0 = 0.2

# Tính tử số: P(F=1, L=1, T=1)
numerator = P_F1 * P_L1_given_F1 * P_T1_given_F1

# Tính mẫu số: P(L=1, T=1)
P_L1_T1 = (
    P_F1 * P_L1_given_F1 * P_T1_given_F1 +
    P_F0 * P_L1_given_F0 * P_T1_given_F0
)

# Bayes: P(F=1 | L=1, T=1)
P_F1_given_L1_T1 = numerator / P_L1_T1

print(f"Xác suất giao dịch là gian lận (P(F=1 | L=1, T=1)): {P_F1_given_L1_T1:.4f}")
```
